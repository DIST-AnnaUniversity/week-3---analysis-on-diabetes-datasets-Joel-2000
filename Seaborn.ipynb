**Seaborn Module**

1.BarPlot

```python
import seaborn as sns
import pandas as pd

d=pd.read_csv('diabetes2.csv')
d=d.iloc[:30]
sns.barplot(y=d['Code'],x=d['Value'])
```

2. LinePlot

```python
import seaborn as sns
import pandas as pd

d=pd.read_csv('diabetes2.csv')
d=d.iloc[:30]
sns.lineplot(y=d['Code'],x=d['Value'])
```

3. lmplot

```python
import seaborn as sns
import pandas as pd

d=pd.read_csv('diabetes2.csv')
d=d.iloc[:30]
sns.lmplot(x="Code",y="Value", data=d)
```

4. Swarmplot

```python
import seaborn as sns
import pandas as pd

d=pd.read_csv('diabetes2.csv')
d=d.iloc[:30]
sns.swarmplot(x="Code",y="Value", data=d)
```
